```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_excel(r'Insurance_fraud_Identification.xlsx')
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Get an overview of the dataset
print(data.info())
```

```python
# Summary statistics for numerical columns
print(data.describe())
```

```python
# Check for missing values
print(data.isnull().sum())
```

```python
# Countplot of 'Month'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Month')
plt.title('Count of Accidents by Month')
plt.show()
```

```python
# Boxplot of 'Age'
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Age')
plt.title('Distribution of Age')
plt.show()
```

```python
# Histogram of 'ClaimSize'
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='ClaimSize', bins=30, kde=True)
plt.title('Distribution of Claim Size')
plt.show()
```

```python
# Pairplot for numerical variables
sns.pairplot(data=data, vars=['Age', 'Deductible', 'ClaimSize'])
plt.title('Pairplot of Numerical Variables')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
